```python
import glob
import os

import tensorflow as tf
import tfx

from tfx.orchestration import metadata
from tfx.types import standard_artifacts
```

```python
metadata_path = os.path.join(os.getcwd(), 'metadata.db')
metadata_connection_config = metadata.sqlite_metadata_connection_config(metadata_path)
```

```python
pipeline_info = tfx.orchestration.data_types.PipelineInfo(pipeline_name='chicago_taxi_pipeline', 
                                                          pipeline_root=os.path.join(os.getcwd(), 'tfx', 'chicago_taxi_pipeline'))

pipeline_info
```

```python
from tfx.types.standard_artifacts import Model, Schema
requested_artifacts = [x.TYPE_NAME for x in [Model, Schema]]

requested_artifacts
```

```python
with metadata.Metadata(metadata_connection_config) as store:
    context = store.get_pipeline_context(pipeline_info)
    artifacts = store.get_published_artifacts_by_type_within_context(requested_artifacts, context.id)

artifacts
```
